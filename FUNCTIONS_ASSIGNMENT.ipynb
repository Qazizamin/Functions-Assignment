### Q5) What are iterators in python and how they differ from iterables?

- An iterator is an object representing a stream of data. It implements the iterator protocol, which consists of two methods: __iter__() and __next__(). Iterators are used to iterate over iterables one element at a time, keeping track of the current state during iteration.
- An iterator is an object returned by the iter() function.
- It maintains internal state to remember the current position during iteration.
- It provides the __next__() method, which returns the next item in the sequence.
- When all items have been returned, it raises a StopIteration exception.



- An iterable in Python is any object capable of returning its members one at a time, allowing it to be iterated over using a loop (e.g., for loop) or other iterative methods. - Iterator: Implements the actual iteration logic, keeping track of the current state.
- Iterable: Provides a way to retrieve an iterator.

- Iterator: Has both __iter__() (which returns itself) and __next__() methods.
- Iterable: Has an __iter__() method that returns an iterator.

- Iterator: Maintains internal state, remembering the current position and providing the next item in the sequence.
- Iterable: Does not maintain state; it's the source of items to iterate over.

### Q6) Explain the concept of generators in python and how they are defined.

- Generators in Python are a powerful way to create iterators. They allow you to generate a sequence of values without storing them in memory. This makes generators very memory efficient and suitable for handling large datasets or generating an indefinite sequence of items.

- Memory Efficiency: Generators do not store all values in memory simultaneously; they generate values dynamically as requested, conserving memory.
- Iteration Support: Generators support the iteration protocol (__iter__() and __next__() methods), making them suitable for use in for loops and other iterable contexts.

### Q5) Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class FibonacciIterator:

    def __init__(self, n):
        self.n = n             # number of terms to generate
        self.current = 0       # current index in the sequence
        self.a, self.b = 0, 1  # initial values for fibonacci sequence

    def __iter__(self):
        return self
    
    def __next__(self):
        if self.current >= self.n:
            raise StopIteration
        else:
            if self.current == 0:
                self.current += 1
                return self.a
            elif self.current == 1:
                self.current += 1
                return self.b
            else:
                self.a, self.b = self.b, self.a + self.b
                self.current += 1
                return self.b
```

```python
fib_iter = FibonacciIterator(15) # Generate Fibonacci sequence up to 15 terms

for number in fib_iter:
    print(number, end = ' ')
```

### Q6) Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def power_of_two(max_exponent):
    exponent = 0

    while exponent <= max_exponent:
        yield 2 ** exponent
        exponent += 1
```

```python
max_exponent = 5  #Generate powers of 2 up to 2^5
for power in power_of_two(max_exponent):
    print(power, end= ' ')
```

### Q7) Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
    """Generator function to read a file line by line."""
    with open(file_path, 'r') as file:
        for line in file:
            yield line.rstrip('\n')  # Remove the newline character
```

```python
file_path = 'sample1.txt'  # Replace with your file path
for line in read_file_line_by_line(file_path):
    print(line)  # Process each line as needed
```
